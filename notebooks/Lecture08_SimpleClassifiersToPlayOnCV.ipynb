## <span style="color:red"> 1. Automatic classification </span> 

The classification is the problem of predicting a discrete class label output for an example.

**Function Approximation:** The idea is  developing a model using historical data to make a prediction on new data where we do not have the answer. This is done by approximating a function:    

$$\hat{f}: x(t) \to y(t)$$

$x(t) = \left \{ x_1, x_2, \ldots x_n  \right \}$ is input data, while $y(t) = \left \{0,1 \right \}$ is output data. For classification, $y(t)$ are discrete output variables. For instance, an e-mail can classified as "spam" and "nor-spam"

```python
import numpy as np
import cv2 as cv
import os
import sys
import tarfile
from skimage import io
from skimage.color import rgb2gray
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from skimage.feature import (match_descriptors, corner_harris,
                             corner_peaks, ORB, plot_matches)
import matplotlib.pyplot as plt
```

```python
path_classes = '../datasets/10_small_ObjectCetegories/'
classes_names = os.listdir(path_classes)
print classes_names
dic_imgs ={}

for i, name_class in enumerate(classes_names):
    dic_imgs[classes_names[i]]=np.array(os.listdir(path_classes + classes_names[i]))
    print "class: ", name_class, " total samples: ", len(dic_imgs[classes_names[i]])
```

## <span style="color:red"> 2. Computing Features </span> 

```python
path_image = path_classes+classes_names[0]+'/'+'image_0001.jpg'
img = io.imread(path_image)
descriptor_extractor = ORB(n_keypoints=200)
descriptor_extractor.detect_and_extract(rgb2gray(img))
keypoints1 = descriptor_extractor.keypoints

plt.figure(figsize=(15,4))
plt.subplot(121);
plt.imshow(img,interpolation="nearest", cmap='gray'); 
plt.subplot(122);
plt.scatter(keypoints1[:,1], keypoints1[:,0],c= "red",edgecolors="black", alpha=.6);
plt.imshow(img,  interpolation="nearest", cmap='gray'); 
```

### Descriptor matrix with and associate label vector

Now, we need to compute a descriptor for each image in whole dataset.
- The images descriptors are coded in a matrix $\mathbf{X}$
    - Each file of the matrix is an image
    - the columns are the ORB descriptors concatenated. 
- The classes of each image if coded into a vector $\mathbf{y}$ 
    - Each element is an integer number that represent the class


```python
descriptor_dataset = ORB(n_keypoints=100)

X =[]
y =[]
for i, classes in enumerate(['garfield','brain', 'car_side' ]):
    print (i+1), "clase: ", classes, " y_value:",  (i+1), " total: ", len(dic_imgs[classes])    
    for im_name in dic_imgs[classes]:
        path_image = path_classes+classes+'/'+im_name
        img = io.imread(path_image)
        descriptor_dataset.detect_and_extract(rgb2gray(img))
        descrip_img = descriptor_dataset.descriptors
        re_des = descrip_img.reshape(descrip_img.shape[0]*descrip_img.shape[1]).astype(int)
        X.append(re_des)
        y.append(i+1)

X = np.r_[X]        
y = np.r_[y]
```

## <span style="color:red"> 3. Linear regressor </span> 

<img src="../imgs/img0.png" style="width:500px;">


- It is the most simple classifer, from which the predicted value $\hat{y}$ is represented as a **linear combination** of input variables $\mathbf{X} = [x_1, x_2, \ldots, x_n]$, so: 

$$\hat{y}(W, X) = w_0 + x_1 w_1+ x_2 w_2+ \ldots + x_n w_n$$
    -(x_i) Inputs: image features, pixels  
    -(W) Weights associated to each of the fetures
    -($\hat{y}$) predictions label. {cat, garfiel, car ...} 
     
**How?:** 

- Fit a linear model with coefficients $w = (w_0, \ldots, w_p)$ minimizing the residual sum of squares: 

$$min_{w} ||\mathbf{X}w -y ||_{2}^{2}$$

    - A natural way to measure the distance between two probability vectors is the cross-entropy.

$$D(S, L) = - \sum_{i} L_{i} \log(S_{i})$$
        
   -$L$ is the vector of labels, and $S$ is the output for some linear combination

- **Solution**: the possible outcomes of a single trial are modeled using a **logistic function**. For instance the SoftMax function. 

$$S(y_i) = \frac{e^{y_i}}{\sum e^{y_{i}}}$$

**Go to the action!**

```python
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=test_size)
print y_test
```

Characterize images with RoB features and classify

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train, y_train)


print clf.score(X_test, y_test)
```

Other way to visualize the results

```python
predicciones_train = clf.predict(X_train)
predicciones_test  = clf.predict(X_test)

cm_train=confusion_matrix(y_train, predicciones_train)
cm_test = confusion_matrix(y_test, predicciones_test)
print cm_train
print "cm_test"
print cm_test
```
